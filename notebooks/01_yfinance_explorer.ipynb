# Exploring yfinance and moving avg

## Download the data from yfinance for stock into dataframe

```python
import yfinance as yf
ticker = 'NAUKRI.NS'
# ticker = 'AAPL'
df = yf.download(ticker, period="1y")
print(df.head())
```

## Smooting the close signal for computer

```python
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA20'].isna()
```

```python
df['MA50'] = df['Close'].rolling(window=50).mean()
```

```python
df.tail()
```

## Plotting the close, MA20, MA50 signal to see the trends

```python
import matplotlib.pyplot as plt
df[['Close', 'MA20', 'MA50']].plot()
plt.show()
```
